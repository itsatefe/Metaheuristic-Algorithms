```python
folder_path='InstanceSets/InstanceSet1/'
instance_name = "cl_01_020_09"
plot_path = 'Plot/' + instance_name
```

```python
def execute_algorithm(dataset, best_value):
    print(f"- dataset: {dataset}")
    bin_width, bin_height, rectangles = read_items(folder_path + dataset)
    optimizer = IslandPSO(config["num_particles"],config["num_islands"],config["max_iteration"], config["island_iteration"], rectangles, config["w"], config["c1"], config["c2"], bin_width, bin_height,
                          config["migration_interval"],config["no_improvement"], config["max_duration"], config["stop_by_time"], best_value=best_value)
#     optimizer = PSO(config["num_particles"],config["max_iteration"], rectangles, config["w"], config["c1"], config["c2"], bin_width, bin_height)
    optimizer.run()
    print(f"- min bins: {best_value}")
    print(f"- found bins: {len(optimizer.global_best.positions)}")
    visualize_bins(optimizer.global_best.positions, 'Plot/'+dataset, 1)
    plot_objectives(optimizer.best_objectives, optimizer.avg_objectives, 'Plot/'+dataset)
    plot_detailed_objectives(optimizer.detailed_objectives,  'Plot/'+dataset)
    # diversity_heatmap(optimizer.diversity, 'Plot/'+dataset)
    print("-"*30)
    return optimizer
```

```python
config = read_config('Config/PSO.json')
datasets = read_excel()
# datasets = [instance_name]
best_values = read_excel_best_solutions()
for i in range(1): 
    bins_size = []
    execution_times = []
    for j, dataset in enumerate(datasets):
        optimizer = execute_algorithm(dataset, best_values[j])
        bins_size.append(len(optimizer.global_best.positions))
        execution_times.append(optimizer.execution_time)
    write_excel_time(execution_times)
    write_excel_bins(bins_size)
    print(f" ------------------------- run number: {i + 1} ----------------------------")
```
